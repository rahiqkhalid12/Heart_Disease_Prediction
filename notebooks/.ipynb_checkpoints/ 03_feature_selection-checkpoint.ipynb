```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import sys

sys.path.append('../utils')
from data_prep import prepare_data

X, y = prepare_data('../data/Heart_Disease.csv')

# Random Forest Classifier model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

rf_model.fit(X, y)

feature_importances = rf_model.feature_importances_

# Get the names of the features
feature_names = X.columns

# Sort the features by importance
sorted_indices = feature_importances.argsort()
sorted_importances = feature_importances[sorted_indices]
sorted_names = feature_names[sorted_indices]

# Create a horizontal bar plot of feature importance
plt.figure(figsize=(12, 8))
plt.barh(range(len(sorted_importances)), sorted_importances, align='center')
plt.yticks(range(len(sorted_names)), sorted_names)
plt.xlabel('Feature Importance')
plt.title('Random Forest Feature Importance')
plt.show()

# Print the top 10 most important features
print("Top 10 Most Important Features:")
for i in sorted_indices[-10:][::-1]:
    print(f"{X.columns[i]}: {feature_importances[i]:.4f}")
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

y_binary = y.apply(lambda x: 0 if x == 0 else 1)

# Use Random Forest as the estimator for RFE
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Create the RFE object and select the top 10 features
rfe = RFE(estimator=rf_model, n_features_to_select=10)
rfe.fit(X, y_binary)

# Get the selected features
selected_rfe_features = X.columns[rfe.support_]
print("Features selected by RFE:")
print(selected_rfe_features)

# Create the reduced dataset
X_rfe = X[selected_rfe_features]
```

```python
from sklearn.feature_selection import chi2, SelectKBest


y_binary = y.apply(lambda x: 0 if x == 0 else 1)

categorical_features = X.select_dtypes(include=['bool', 'int']).columns.tolist()

# Select the top 10 features using Chi-Square
chi2_selector = SelectKBest(chi2, k=10)
chi2_selector.fit(X[categorical_features], y_binary)

# Get the selected features
selected_chi2_features = X[categorical_features].columns[chi2_selector.get_support()]
print("Features selected by Chi-Square test:")
print(selected_chi2_features)
```

```python
top_rf_features = X.columns[sorted_indices[-10:]]

# Create the final reduced dataset for modeling
X_reduced = X[top_rf_features]

# Print the final shape to confirm the reduction
print("Shape of the original dataset:", X.shape)
print("Shape of the reduced dataset:", X_reduced.shape)
```
