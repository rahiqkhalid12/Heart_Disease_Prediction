```python
import pandas as pd
import numpy as np

# Load the dataset with the correct filename from your folder
df = pd.read_csv('../data/heart_disease.csv')

# Display the first 5 rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())

# Get a summary of the dataset
print("\nDataset Information:")
df.info()

# Get descriptive statistics of the numerical columns
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
# Check for the number of missing values in each column
print("Missing values before imputation:")
print(df.isnull().sum())

# Impute missing values in 'ca' and 'thal' with the median of each column
df['ca'] = df['ca'].fillna(df['ca'].median())
df['thal'] = df['thal'].fillna(df['thal'].median())

# Check again to confirm there are no more missing values
print("\nMissing values after imputation:")
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import StandardScaler
import sys
sys.path.append('../utils')
from data_prep import prepare_data

X, y = prepare_data('../data/Heart_Disease.csv')

# Display the first 5 rows of the new, preprocessed dataframe
print("First 5 rows of the preprocessed dataset:")
print(X.head())

# Display the new column information to see the result of encoding
print("\nPreprocessed Dataset Information:")
X.info()
```

```python
import matplotlib.pyplot as plt

numerical_features = X.select_dtypes(include=['float64']).columns
X[numerical_features].hist(bins=15, figsize=(15, 10), layout=(3, 3))
plt.suptitle('Histograms of Numerical Features')
plt.show()
```

```python
import seaborn as sns

y_binary = y.apply(lambda x: 0 if x == 0 else 1)
preprocessed_df = pd.concat([X, y_binary], axis=1)

# Calculate the correlation matrix
correlation_matrix = preprocessed_df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap of All Features')
plt.show()
```

```python
numerical_features = X.select_dtypes(include=['float64']).columns

# Create boxplots
plt.figure(figsize=(15, 10))
X[numerical_features].boxplot()
plt.title('Boxplots of Numerical Features')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
